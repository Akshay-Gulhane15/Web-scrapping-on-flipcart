```python
prod_name = []
prod_price = []
prod_rating = []
prod_no_of_reviews = []
for i in master_container:
    name = i.find('a',{'class':'s1Q9rs'})
    prod_name.append(name)
    price = i.find('div',{'class':'_30jeq3'})
    prod_price.append(price)
    rating = i.find('div',{'class':'_3LWZlK'})
    prod_rating.append(rating)
    no_of_reviews=i.find('span',{'class':'_2_R_DZ'})
    prod_no_of_reviews.append(no_of_reviews)
    
```

```python
final_df = pd.DataFrame({'Product Name':prod_name,
                        'Product Price':prod_price,
                        'Product no of reviews': prod_no_of_reviews,
                        'Product Rating':prod_rating})

final_df
```

```python
prod_name = []
prod_rating = []
prod_price = []
prod_no_of_reviews=[]
#The below line would iterate the number for every page
for i in range(1,11):
    url = 'https://www.flipkart.com/search?q=books&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page='+str(i)
    #We need to get access to the url and get the html
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'lxml')
    #Getting the master container so that we can get the data
    master_container = soup.find_all('div',{'class':'_4ddWXP'})
    for j in master_container:
        name = j.find('a',{'class':'s1Q9rs'}).text
        prod_name.append(name)
        # Price
        price = j.find('div',{'class':'_30jeq3'})
        if price is not None:
            prod_price.append(price.text)
        else:
            prod_price.append("NA")
        # Rattings   
        rating = j.find('div',{'class':'_3LWZlK'})
        if rating is not None:
            prod_rating.append(rating.text)
        else:
            prod_rating.append('NA')
        # Retail Price  
        no_of_reviews = j.find('span',{'class':'_2_R_DZ'})
        if no_of_reviews is not None:
            prod_no_of_reviews.append(no_of_reviews.text)
        else:
            prod_no_of_reviews.append('NA')
```

```python
len(prod_name),len(prod_rating),len(prod_price),len(prod_no_of_reviews)
```

```python
final_df = pd.DataFrame({'Product Name':prod_name,
                        'Product Price':prod_price,
                        'Product no of reviews': prod_no_of_reviews,
                        'Product Rating':prod_rating})

final_df
```
